## 1. Python Data Structures
We'll start by defining some basic structures to hold student data.

```python
# List of student names
students = ["Alice", "Bob", "Charlie", "David", "Eva"]

# Dictionary for a single student profile
student_profile = {
    "name": "Alice",
    "id": 101,
    "courses": ["Math", "Science"]
}

# Tuple (Immutable sequence)
coordinates = (31.5204, 74.3587) # Latitude and Longitude of Lahore

# Set (Unique elements)
unique_subjects = {"Math", "Science", "Math", "English"}

print(f"Students: {students}")
print(f"Profile: {student_profile}")
print(f"Unique Subjects: {unique_subjects}")
```

## 2. File Processing (CSV and JSON)
Processing actual dataset files.

```python
import csv
import json

def process_csv(file_path):
    """Reads student grades from a CSV file."""
    data = []
    try:
        with open(file_path, mode='r') as file:
            reader = csv.DictReader(file)
            for row in reader:
                # Data transformation: String to Int
                row['math'] = int(row['math'])
                row['science'] = int(row['science'])
                row['english'] = int(row['english'])
                data.append(row)
    except FileNotFoundError:
        print("Error: CSV file not found.")
    except Exception as e:
        print(f"An error occurred: {e}")
    return data

def process_json(file_path):
    """Reads student grades from a JSON file."""
    try:
        with open(file_path, mode='r') as file:
            return json.load(file)
    except Exception as e:
        print(f"Error reading JSON: {e}")
        return []

csv_data = process_csv('students.csv')
json_data = process_json('students.json')

print(f"Loaded {len(csv_data)} records from CSV.")
print(f"First record: {csv_data[0] if csv_data else 'None'}")
```

## 3. Data Transformation with List Comprehensions
Efficiently calculating averages and identifying top students.

```python
# Calculate average for each student using list comprehension
student_averages = [
    {
        "name": s['name'], 
        "avg": (s['math'] + s['science'] + s['english']) / 3
    } 
    for s in csv_data
]

# Filter students with average > 80
high_achievers = [s for s in student_averages if s['avg'] > 80]

print("High Achievers:")
for s in high_achievers:
    print(f"{s['name']}: {s['avg']:.2f}")
```

## 4. Error Handling and writing to Text File
Generating a report while handling potential errors.

```python
def save_report(data, output_file):
    try:
        with open(output_file, 'w') as f:
            f.write("STUDENT PERFORMANCE REPORT\n")
            f.write("=" * 30 + "\n")
            for item in data:
                f.write(f"{item['name']}: {item['avg']:.2f}\n")
        print(f"Report saved to {output_file}")
    except IOError as e:
        print(f"Could not write to file: {e}")

save_report(student_averages, 'grades_report.txt')
```

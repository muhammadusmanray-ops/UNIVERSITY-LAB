This notebook covers essential data manipulation and visualization tasks using the Titanic dataset.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Task 18: Load data from Seaborn library
df = sns.load_dataset('titanic')

# Task 19: Initial Exploration
print("Total Rows & Columns:", df.shape)
df.head()
```

```python
# Task 20: Select columns and filter rows using Boolean indexing
# Filter for Female Survivors
survivors = df[(df['survived'] == 1) & (df['sex'] == 'female')]
print("Total Female Survivors:", len(survivors))
survivors.head()
```

```python
# Task 21: Sort data by single and multiple columns
# Sort by Fare (Descending)
most_expensive = df.sort_values(by='fare', ascending=False)
print("Top 5 Most Expensive Tickets:\n", most_expensive[['pclass', 'fare', 'survived']].head())

# Sort by Class and Age (Ascending)
sorted_multi = df.sort_values(by=['pclass', 'age'], ascending=[True, True])
print("\nSorted by Class and then Age:\n", sorted_multi[['pclass', 'age', 'survived']].head())
```

```python
# Task 22: Handle Missing Data
print("Missing values in Age before:", df['age'].isnull().sum())

# fillna (Average age)
avg_age = df['age'].mean()
df['age'] = df['age'].fillna(avg_age)

# dropna (Embarked row deletion)
df = df.dropna(subset=['embarked'])

print("Ab Embarked mein khali values:", df['embarked'].isnull().sum())
print("Ab Age mein khali values:", df['age'].isnull().sum())
```

```python
# Task 23: Add, Remove, and Rename Columns
# 1. Add Column: family_size
df['family_size'] = df['sibsp'] + df['parch'] + 1

# 2. Rename Column: fare -> Ticket_Price, pclass -> Passenger_Class
df = df.rename(columns={'fare': 'Ticket_Price', 'pclass': 'Passenger_Class'})

# 3. Remove Column
df_final = df.drop(columns=['who', 'adult_male'])

print("Final Cleaned & Manipulated Data ready!")
df_final[['age', 'sex', 'Ticket_Price', 'family_size']].head()
```

## Data Visualization
Checking patterns and distributions in the data.

```python
# 1. Survival Count Plot (Kitne bache aur kitne fauth huye)
plt.figure(figsize=(8, 5))
sns.countplot(data=df, x='survived', hue='sex')
plt.title("Survivors vs Non-Survivors by Gender")
plt.show()

# 2. Age Distribution (Kaunsi age ke log zyada thay?)
plt.figure(figsize=(8, 5))
sns.histplot(df['age'], bins=20, kde=True, color='green')
plt.title("Passenger Age Distribution")
plt.show()
```
